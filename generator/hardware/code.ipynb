At last, we define the function to split up the paths into multiple blocks. We firstly define the euclidean distance function to calculate the euclidian distance between two vertices.

```python
def euclidean_distances(i_x, i_y, j_x, j_y):
    return int(np.round(np.sqrt((i_x - j_x) ** 2 + (i_y - j_y) ** 2)))
```

Then we define the function to add the routine string to the given string row. Note that the generated path should be black (to be clearer).

```python
def string_addition_routine(string_row, x_small, x_large, y_small, y_large):
    string_row += "\tif (DrawX >= " + str(x_small) + " && DrawX <= " + str(x_large) + " && DrawY >= "
    string_row += str(y_small) + " && DrawY <= " + str(y_large) + ") // " + points[i[0]][2] + "->" + points[i[1]][2] + "\n"
    string_row += "\tbegin\n"
    string_row += "\t\tRed=8'd0;\n"
    string_row += "\t\tGreen=8'd0;\n"
    string_row += "\t\tBlue=8'd0;\n"
    string_row += "\tend\n"
    return string_row
```

```python
string=''
counter = 0
slope = 0
path_counter = 1

# each edges is a pair of vertices
for i in edges:
    # initialize the row of string
    string_row = 'if (path' + str(path_counter) + '_export['
    string_row += str(counter) + "] == 1)\n"
    string_row += "begin\n"

    # get the x and y coordinates of the pair of the vertices
    x1 = int(points[i[0]][0])
    y1 = int(points[i[0]][1])
    x2 = int(points[i[1]][0])
    y2 = int(points[i[1]][1])
    
    # calculate the relative positions of the pair of vertices
    x_large = max(x1, x2)
    x_small = min(x1, x2)
    y_large = max(y1, y2)
    y_small = min(y1, y2)

    # calculate the slope of the edge in the VGA coordination
    if (x1 == x_small and y1 == y_large) or (x1 == x_large and y1 == y_small):
        slope = -1
    elif (x1 == x_small and y1 == y_small) or (x1 == x_large and y1 == y_large):
        slope = 1
    else:
        slope = 0

    # calculate the number of blocks required to split the line for rendering with step size (block width) of 4
    block_size = 6
    num_block = int(euclidean_distances(x1, y1, x2, y2) / block_size)
    x_intermediate_value_list = np.linspace(x_small, x_large, num=num_block, endpoint=True)
    y_intermediate_value_list = np.linspace(y_small, y_large, num=num_block, endpoint=True)

    # iterate through all the blocks and generate code for each one
    for j in range(num_block-1):

        if slope >= 0:
            string_row = string_addition_routine(string_row, int(x_intermediate_value_list[j]), int(x_intermediate_value_list[j+1]), int(y_intermediate_value_list[j]), int(y_intermediate_value_list[j+1]))
        else:
            string_row = string_addition_routine(string_row, int(x_intermediate_value_list[j]), int(x_intermediate_value_list[j+1]), int(y_intermediate_value_list[num_block-2-j]), int(y_intermediate_value_list[num_block-1-j]))

    string_row += "end\n"
    string += string_row 
    string += "\n"
    counter += 1

    if counter == 32:
        counter = 0
        path_counter += 1

string += "\n"

with open('./render_limit_blocks.txt', "w") as f:
    f.write(string)
```
